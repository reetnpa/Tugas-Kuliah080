```python
bb = int(input("Silahkan masukkan batas bawah : "))
```

```python
ba = int(input("Silahkan masukkan batas atas : "))
```

```python
import math

while bb <= ba :
    akar_bilangan = math.sqrt(bb)
    print(f"akar {bb} = {akar_bilangan:.2f}")
    bb += 1
```
