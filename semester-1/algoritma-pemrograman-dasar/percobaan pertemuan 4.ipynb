```python
# Menentukan karakter vokal atau konsonan
```

```python
print("Sistem ini akan menguji karakter vokal, konsonan atau angka")
ktr = str(input("Masukkan karakter yang ingin di uji : "))
if (ktr == 'A' or ktr == 'I' or
    ktr == 'U' or ktr == 'E' or ktr == 'O'):
    print("karakter vokal")
elif (ktr == '0' or ktr == '1' or
      ktr == '2' or ktr == '3' or ktr == '4'):
    print("karakter angka")
else :
    print("karakter konsonan")
print("Terima kasih telah di uji")
```
