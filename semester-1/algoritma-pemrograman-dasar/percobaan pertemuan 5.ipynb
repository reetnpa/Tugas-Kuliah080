```python
Bulan = int(input("Masukkan nomor urut bulan yang di uji"))
```

```python
if Bulan == 0 :
    print("Bukan nama bulan")
elif Bulan == 1 :
    print("Bulan Januari")
elif Bulan == 2 :
    print("Bulan Februari")
elif Bulan == 3 :
    print("Bulan Maret")
elif Bulan == 4 :
    print("Bulan April")
elif Bulan == 5 :
    print("Bulan Mei")
elif Bulan == 6 :
    print("Bulan Juni")
elif Bulan == 7 :
    print("Bulan Juli")
elif Bulan == 8 :
    print("Bulan Agustus")
elif Bulan == 9 :
    print("Bulan September")
elif Bulan == 10 :
    print("Bulan Oktober")
elif Bulan == 11 :
    print("Bulan November")
elif Bulan == 12 :
    print("Bulan Desember")
else :
    print("Tidak tersedia")
print("Pengujian selesai")
```
