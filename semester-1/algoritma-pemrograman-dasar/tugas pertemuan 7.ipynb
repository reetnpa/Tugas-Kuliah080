```python
from goto import with_goto

@with_goto
def input_bulan():
  label .inputbulan
  bulan = int(input("Masukkan nama bulan : "))
    
  if bulan == 0 :
    for x bulan in ['Januari', 'Februari', 'Maret']:
      if (bulan) :
          print("Termasuk: Quarter 1")
      else :
          break 
          
          goto .inputbulan

  elif bulan in ['April', 'Mei', 'Juni']:
    print("Termasuk: Quarter 2")
  elif bulan in ['Juli', 'Agustus', 'September']:
    print("Termasuk: Quarter 3")
  elif bulan in ['Oktober', 'November', 'Desember']:
    print("Termasuk: Quarter 4")
  else:
    print("Maaf, bulan tidak terdeteksi") 

  input_bulan()
```

```python
class GotoException(Exception):
    pass

def input_bulan(bulan):
    try:
        if bulan in ['Januari', 'Februari', 'Maret']:
            raise GotoException(f"{bulan} Termasuk: Quarter 1.")
        elif bulan in ['April', 'Mei', 'Juni']:
            raise GotoException(f"{bulan} Termasuk: Quarter 2.")
        elif bulan in ['Juli', 'Agustus', 'September']:
            raise GotoException(f"{bulan} Termasuk: Quarter 3.")
        elif bulan in ['Oktober', 'November', 'Desember']:
            raise GotoException(f"{bulan} Termasuk: Quarter 4.")
        else:
            print("Maaf, bulan tidak terdeteksi")
            return False 
    except GotoException as x :
        print(x)
        return True

while True :
    bulan = str(input("Masukkan nama bulan : "))
    if input_bulan(bulan):
        break
```

```python
tinggi = eval(input("Masukkan tinggi piramida bintang : "))
for x in range(tinggi) :
    for count in range(tinggi-x) :
        print(end=" ")
    for count in range(x+1) :
        print(end="* ")
    print()
```

```python
rows = 5

for i in range(rows-1, -1, -1):
    for j in range(rows -i -1):
        print(" ", end="")
    for j in range(2*i+1):
        print("*", end="")
    print()
```
