```python
jumlah = 0

# Request input from the user
print("Masukkan operasi aritmatika 2 bilangan, jika hasilnya negatif maka akan berakhir : ")
while True : # Loop forever
    entry = eval(input()) # Get the vakue
    if entry < 0 : # jika hasil = -1,-2,-3,...
        break # entry < 0 = stop
    jumlah = jumlah + entry

print("Sum = ", jumlah)
```

```python
kolom = int(input("Jumlah kolom : "))
```

```python
for m in range(1,baris+1) :
    for n in range(1,kolom+1):
        hasil_kali=m*n
        print(m,"*",n,":",hasil_kali)
    print("\n")
```
