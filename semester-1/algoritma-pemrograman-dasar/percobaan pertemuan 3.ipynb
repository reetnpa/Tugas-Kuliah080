```python
# konversi rupiah ke dollar USD
```

```python
# input nilai rupiah
rupiah = int(input("Masukkan nilai rupiah:"))
```

```python
# input nilai dollar
dollar = int(input("Masukkan nilai dollar:"))
```

```python
# input hasil
hasil = rupiah/dollar
```

```python
print("hasil konversi= %.2f dollar USD" % hasil)
```

```python
# luas trapesium
```

```python
# input nilai sisi atas (a)
a = int(input("Masukkan nilai panjang sisi atas CM:"))
```

```python
# input nilai sisi bawah (b)
b = int(input("Masukkan nilai panjang sisi bawah CM:"))
```

```python
# input nilai sisi tinggi (t)
t = int(input("Masukkan nilai panjang sisi tinggi CM:"))
```

```python
# rumus luas trapesium
luas = 1/2*(a+b)*t
```

```python
print("hasil luas trapesium= %d CM" % luas)
```

```python
# Rata-rata nilai ujian siswa 
```

```python
# input nilai siswa 1
Retno =int(input("Masukkan nilai hasil ujian Retno:"))
```

```python
# input nilai siswa 2
Nesa =int(input("Masukkan nilai hasil ujian Nesa:"))
```

```python
# input nilai siswa 3
Sashi =int(input("Masukkan nilai hasil ujian Sashi:"))
```

```python
# input nilai siswa 4
Andin =int(input("Masukkan nilai hasil ujian Andin:"))
```

```python
# input nilai siswa 5
Zahra =int(input("Masukkan nilai hasil ujian Zahra:"))
```

```python
# menghitung rata-rata
hasil = (Retno+Nesa+Sashi+Andin+Zahra)/5
```

```python
print("hasil rata-rata nilai ujian kelima siswa= %d" % hasil)
```
