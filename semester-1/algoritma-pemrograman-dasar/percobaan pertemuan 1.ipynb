```python
# iteration using dictionary

params = {"param1" : 1.0,
          "param2" : 2.0,
          "param3" : 3.0}

for key in params:
    print(key + ": " + str(params[key]))

print("")
print("enumerate:")

for key,param in enumerate(params):
    print(key)
    print(param)
```
