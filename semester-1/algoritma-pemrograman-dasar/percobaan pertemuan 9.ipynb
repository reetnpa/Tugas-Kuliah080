```python
import time
from IPython.display import clear_output

# Jumlah paragraf dan langkah dalam satu paragraf
paragraphs = 3
steps_per_paragraph = 15

# Loop untuk setiap paragraf
for paragraph in range(paragraphs):
    # Loop untuk setiap langkah dalam paragraf
    for step in range(steps_per_paragraph):
        clear_output()
        # Cetak bintang pada posisi saat ini
        print(" " * step + "*")
        time.sleep(0.8)
    # Pindah ke paragraf baru setelah selesai satu paragraf
    print()  # Dua baris kosong antara paragraf

# Cetak baris terakhir dengan "Hello World!" setelah bintang mencapai posisi akhir
clear_output()
print(" " * steps_per_paragraph + "* Hello World!")
```

```python
import time
from IPython.display import clear_output

# Jumlah langkah dalam satu baris dan jumlah baris yang diinginkan
steps_per_line = 15
lines = 3

# Loop untuk setiap baris
for line in range(lines):
    # Loop untuk setiap langkah dalam baris
    for step in range(steps_per_line):
        clear_output(wait=True)
        # Cetak bintang dengan posisi horizontal dan vertikal sesuai saat ini
        print("\n" * line + " " * step + "*")
        time.sleep(0.5)

# Cetak baris terakhir dengan "Hello World!" setelah bintang mencapai posisi akhir
clear_output(wait=True)
print("\n" * lines + " " * steps_per_line + "* Hello World!")
```
