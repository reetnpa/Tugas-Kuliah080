```python
df['genres_split'] = df['genres'].str.split(', ')
df_exploded = df.explode('genres_split')
rating_by_genre = df_exploded.groupby('genres_split')['imdbAverageRating'].mean().reset_index()

plt.figure(figsize=(18, 9))
plt.plot(rating_by_genre['genres_split'], rating_by_genre['imdbAverageRating'], marker='o', color='b', linewidth=2)
plt.title("Average IMDb Rating by Genre")
plt.xlabel("Genre")
plt.ylabel("Average IMDb Rating")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
rating_by_type = df.groupby('type')['imdbAverageRating'].mean().reset_index()

plt.figure(figsize=(8, 5))
plt.plot(rating_by_type['type'], rating_by_type['imdbAverageRating'], marker='o', color='g', linewidth=2)
plt.title("Average IMDb Rating by Type")
plt.xlabel("Type")
plt.ylabel("Average IMDb Rating")
plt.grid(True)
plt.show()
```

```python
genre_type_counts = df_exploded.groupby(['genres_split', 'type']).size().unstack(fill_value=0)

genre_type_counts.plot(kind='line', figsize=(12, 6), marker='o')
plt.title("Genre Distribution by Content Type")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

df = pd.read_csv('dataset_bersih.csv')
if 'title' in df.columns:
    vectorizer = CountVectorizer().fit_transform(df['title'])
    cosine_sim = cosine_similarity(vectorizer)

    film_title = "Divergent"

    try:
        idx = df[df['title'] == film_title].index[0]
        sim_scores = list(enumerate(cosine_sim[idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        top_similar = sim_scores[1:6]
        recommended_movies = [df['title'].iloc[i[0]] for i in top_similar]

        print("Rekomendasi film:")
        for movie in recommended_movies:
            print(movie)

    except IndexError:
        print(f"Film dengan judul '{film_title}' tidak ditemukan di dataset.")
else:
    print("Kolom 'title' tidak ditemukan di dataset.")
```
